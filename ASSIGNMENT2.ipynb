```python
import pandas as pd
from IPython.display import display 


df = pd.read_csv("amazon.csv")  

missing_mask = df.isnull().any(axis=1)


print("Rows with missing values before filling:")
display(df[missing_mask])


for column in df.columns:
    mode = df[column].mode()[0]  
    df[column].fillna(mode, inplace=True)  

print("\nRows with missing values after filling:")
display(df[df.isnull().any(axis=1)])


print("\nFirst few rows after filling:")
display(df.head())


print("\nEntire filled dataframe:")
display(df)
```

QUESTION 2  DATA CLEANING

```python
import pandas as pd
import re  


df = pd.read_csv("amazon.csv")


df['discounted_price'] = df['discounted_price'].str.replace('[‚Çπ$,]', '', regex=True)
df['actual_price'] = df['actual_price'].str.replace('[‚Çπ$,]', '', regex=True)


df['discounted_price'] = pd.to_numeric(df['discounted_price'], errors='coerce')
df['actual_price'] = pd.to_numeric(df['actual_price'], errors='coerce')


df['rating_count'] = df['rating_count'].str.replace('[^0-9-]', '', regex=True)


df['rating_count'] = pd.to_numeric(df['rating_count'], errors='coerce').fillna(0)


df['rating_count'] = df['rating_count'].astype(int)


print(df.dtypes)


df.to_csv("amazon_cleaned.csv", index=False)

print("Price and rating_count conversions completed. Cleaned data saved to amazon_cleaned.csv")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sqlalchemy import create_engine
from sqlalchemy.exc import SQLAlchemyError


user = 'root'
password = '130101600Judy!'
host = 'localhost'
database = 'ecommerce_db'

engine = create_engine(f'mysql+pymysql://{user}:{password}@{host}/{database}')

sql_query = """
SELECT category, SUM(rating_count) AS total_reviews
FROM products
GROUP BY category
ORDER BY total_reviews DESC;
"""

try:
    df = pd.read_sql(sql_query, con=engine)
    print("Data loaded successfully.")
except SQLAlchemyError as e:
    print(f"Error loading data: {e}")
    df = pd.DataFrame()  
```

```python
if df.empty:
    print("DataFrame is empty. Cannot create pie chart.")
else:
    
    plt.figure(figsize=(10, 10))
    plt.pie(df['total_reviews'], labels=df['category'], autopct='%1.1f%%', startangle=140)
    plt.title('Top-Selling Categories (Based on Reviews)')
    plt.axis('equal')  
    plt.show()
```

Potential Anomalies:

Outliers in Ratings:
Extremely High or Low Ratings: Products with unusually high (e.g., 5.0) or low (e.g., 1.0) ratings might indicate:
Review manipulation (e.g., fake reviews).
Exceptional product quality or severe defects.
Errors in data entry.
How to Investigate:
Examine the number of reviews associated with these outliers.
Analyze the text of the reviews (if available) for suspicious patterns.
Check for consistency with other product metrics.
Discrepancies in Pricing:
Negative Discount Amounts: This is a clear data error.
Extremely High Discounts: Discounts exceeding a realistic threshold (e.g., 90%) might be:
Data entry errors.
Promotional offers for a limited time.
Clearance items.
How to Investigate:
Check the data source for accuracy.
Compare prices with other online retailers.
Inconsistencies in Review Counts:
